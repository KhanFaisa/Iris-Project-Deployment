### Step 1 : Read The Dataset

```python
import pandas as pd 
df = pd.read_csv('iris.csv')
df.head()
```

### Step 2 : Perform Basic Data Quality Checks 

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3 : Seprate X & Y

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind='bar')
```

### Step 4 : Create A Preprocessing Pipeline For X & Y

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

```python
num_pipe
```

### Save Above num_pipe

```python
import pickle
```

```python
with open('pre.pkl', 'wb') as file1:
    pickle.dump(num_pipe, file1)
```

### Step 5 : Train Test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6 : Build & Save The Model 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

### Step 7 : Evaluate The Model 


```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Plot The Confusion Matrix For Test

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

### print The Classification Report

```python
ypred_train = model.predict(xtrain)
ypred_test = model.predict(xtest)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

### Get Cross Validated f1_macro Score

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring='f1_macro')
```

```python
scores
```

```python
scores.mean()
```

### Above model has f1_macro 5 fold cross validated score of 0.9319 > 0.8 hence above is a good model
